## Object Detection Using YOLO Project- YOLO Model Training

### Download the generated dataset from Roboflow where objects in images are identified, label annotated and all images are resized into 640 x 640 to fed into YOLO model.

```python
from roboflow import Roboflow
from ultralytics import YOLO

# 1. Download your dataset from Roboflow
rf = Roboflow(api_key="j477ebLsOvfw3f2TlyW6")
project = rf.workspace("saranyas-workspace-wv1kd").project("object-detection-yolo-7td8i")
version = project.version(2)
dataset = version.download("yolo26") # This works for YOLOv8, YOLOv11, and YOLO26

# The variable 'dataset.location' now points to the folder containing your 'data.yaml'
```

#### Train the Model

```python
model = YOLO('yolo26n.pt') 

# 3. Start Training
# 'data' points to the .yaml file that Roboflow generated for you
results = model.train(
    data=f"{dataset.location}/data.yaml", 
    epochs=50,      # Number of times the model sees the entire dataset
    imgsz=640,     # Standard resolution for YOLO
    plots=True     # Generates charts so you can monitor progress
)
```
